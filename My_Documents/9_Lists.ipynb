Right : y, u, i, o, p, h, j, k, l, n, m

```python
word = input('Enter a word:')
right = 'yuiophjklnm'
left = 'qwertasdfgzxcvb'

if set(word).intersection(set(right)) and set(word).intersection(set(left)) :
  print('True')
else :
  print('False')
```
