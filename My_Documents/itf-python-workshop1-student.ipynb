### Problem 14

Write a Python program to compute average of two given lists.

nums1 = [1, 1, 3, 4, 4, 5, 6, 7]

```python
nums1 = [1, 1, 3, 4, 4, 5, 6, 7] 
nums2 = [0, 1, 2, 3, 4, 4, 5, 7, 8]
result = (sum(nums1) + sum(nums2)) / len(nums1) + len(nums2)

print(result)
import statistics
statistics.mean(nums1 + nums2)
```
